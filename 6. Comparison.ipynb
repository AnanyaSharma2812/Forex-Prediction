```python
import sys
import os
import warnings

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
if not sys.warnoptions:
    warnings.simplefilter("ignore")
    os.environ["PYTHONWARNINGS"] = "ignore" # Also affect subprocesses

# set style of charts
sns.set(style="darkgrid")
plt.rcParams['figure.figsize'] = [10, 10]
```

```python
model_scores = pd.read_csv("./Model scores.csv", index_col=0)
model_scores
```

```python
# Create bar chart for to show MAE of all models side by side
plt.figure(figsize=(10,6))
sns.barplot(x="MAE", y = "Model", data=model_scores, orient='h', order=model_scores.sort_values('MAE').Model)
plt.xlabel('Mean Absolute Error', fontsize=15)
plt.xticks(rotation=45, fontsize=15)
plt.title('Mean Absolute Error of All Models Tested', fontsize=18)
plt.ylabel("Model", fontsize=15)
plt.yticks(fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# Create bar chart for to show MAE of all models side by side
plt.figure(figsize=(10,6))
sns.barplot(x="MAPE in %", y = "Model", data=model_scores, orient='h', order=model_scores.sort_values('MAPE in %').Model)
plt.xlabel('Mean Absolute Percentage Error in %', fontsize=15)
plt.xticks(rotation=45, fontsize=15)
plt.title('Mean Absolute Percentage Error of All Models Tested', fontsize=18)
plt.ylabel("Model", fontsize=15)
plt.yticks(fontsize=15)
plt.tight_layout()
plt.show()
```
